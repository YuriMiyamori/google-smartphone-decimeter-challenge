```julia
function get_gdf_by_key(gd::GroupedDataFrame, key::Tuple{Any})::Union{DataFrame, Missing}
    if haskey(gd.keymap, key)
        num_gdf = gd.keymap[key]
        return gd[num_gdf]
    else
        return missing
    end
end
```

```julia
function dist2(obs_x::F, obs_y::F, gt_x::Vector{F}, gt_y::Vector{F})::Vector{F}
    return  (obs_x .- gt_x).^2 + (obs_y .- gt_y).^2
end
function nearest_points(obs_x::Vector{F}, obs_y::Vector{F}, gt_x::Vector{F}, gt_y::Vector{F})::Tuple{Vector{F}, Vector{F}}
    n = length(obs_x)
    nearest_x = similar(obs_x, F)
    nearest_y = similar(obs_y, F)
    for i in 1:n
        d2 = dist2(obs_x[i], obs_y[i], gt_x, gt_y)
        min_idx = findmin(d2)[2]
        nearest_x[i] = gt_x[min_idx]
        nearest_y[i] = gt_y[min_idx]
    end
    return nearest_x, nearest_y
end
function nearest_points(df::DataFrame, obs_x::S, obs_y::S)
    out_x = S("$(obs_x)_nearest_gt")
    out_y = S("$(obs_y)_nearest_gt")
    df[!, out_x] = similar(df[!, obs_x] ,F)
    df[!, out_y] = similar(df[!, obs_y], F)
    gd = groupby(df, :phone)
    for (gdf, key) in zip(gd, keys(gd))
        x, y = nearest_points(Array(gdf[:, obs_x]), Array(gdf[:, obs_y]), Array(gt_df[:, :lngDeg_gt]), Array(gt_df[:, :latDeg_gt]))
        gdf[:, out_x] = x
        gdf[:, out_y] = y
    end
end

nearest_points(train_df_concat_sjc, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean, :latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean)
nearest_points(test_df_concat_sjc, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean, :latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean)
```

```julia
function main(df1, df2)
    df = leftjoin(df1, @select(df2, :phone, :time_UTC,
            :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt, :latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt), on=[ :phone, :time_UTC])
    for i in 1:size(df, 1)
        if ismissing(df[i, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt])
            df[i, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt] = df[i, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean] 
            df[i, :latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt] = df[i, :latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean] 
        end
    end
    df.lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt = F.(df.lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt)
    df.latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt = F.(df.latDeg_Y_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt);
    # util_fc.lnglat2XY(df, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt, :lngDeg_X_obs_rjct_outlier_klmn_mix_stop_mean_nearest_gt)
    return df
end

train_df_concat = main(train_df_concat, train_df_concat_sjc) 
test_df_concat = main(test_df_concat, test_df_concat_sjc) 
```
